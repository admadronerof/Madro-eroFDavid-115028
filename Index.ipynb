```python
#Taller punto 14

texto=(input('escriba un texto: '))
#El comando sorted nos ayuda para devolver todo los elementos iterables de la
#lista en orden asendente 
texto1=sorted(texto)
texto2=[]
#esta concion me es util para establecer el texto1 en otro texto2 y volverlo
#una lista de forma horizontal.
for i in range(0,len(texto1)): #se establece el rango para la variable i
    if texto1[i]!=' ':
        
        texto2.append(texto1[i])
        
print(texto2)

cadena = "".join(texto2)

print(cadena)
```

```python
#Taller punto 15
Entero1=(int(input('digite el número de elemntos que desea: '))) #cantidad num.
V=[]            #vector con numeros enteros
                #aqui se pide los numeros enteros y se guardan en la lista
for i in range(Entero1):
    V.append(input('digite el numero'+str(i+1)+':'))
#aqui se pregunta las veces que esta el numero en vec[i] en su mismo vector
#en caso de que se encuentre dos veces se imprime que se repite 2 veces
lista=[]
for i in range(len(V)):
    X=V.count(V[i])
    if (X==2):
        cadena='numero '+V[i]+'su número se repite 2 veces en el vector'
        if cadena not in lista:
            lista.append(cadena)
            
for i in range(len(lista)):
    print(lista[i])
```
